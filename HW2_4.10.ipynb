## 4.10  
画出下列每个 ARMA 模型的自相关函数：

### (a) 
ARMA(1, 1)，$\phi = 0.7$，$\theta = 0.4$

### (b) 
ARMA(1, 1)，$\phi = 0.7$，$\theta = -0.4$

```r
# 加载必要的库
library(stats)

# 定义ARMA模型的参数
phi_a <- 0.7
theta_a <- 0.4
phi_b <- 0.7
theta_b <- -0.4

# 生成ARMA(1,1)模型的理论自相关函数
acf_a <- ARMAacf(ar = phi_a, ma = theta_a, lag.max = 20)
acf_b <- ARMAacf(ar = phi_b, ma = theta_b, lag.max = 20)

# 绘制自相关函数图
par(mfrow = c(2, 1)) # 设置绘图区域为2行1列

# 绘制第一个模型的自相关函数
acf(acf_a, main = expression((a):~ARMA(1, 1)~with~phi == 0.7, theta == 0.4), xlab = "Lag", ylab = "ACF")

# 绘制第二个模型的自相关函数
acf(acf_b, main = expression((b):~ARMA(1, 1)~with~phi == 0.7, theta == -0.4), xlab = "Lag", ylab = "ACF")
```
